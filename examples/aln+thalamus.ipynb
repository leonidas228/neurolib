```python
from neurolib.models.aln_thalamus import ALNThalamusModel
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (20, 9)
```

```python
cmat = np.random.rand(2, 2)
dmat = np.random.rand(2, 2)

n = ALNThalamusModel(Cmat=cmat, Dmat=dmat, thlm_cmat=np.array([0.1, 0.1]), thlm_dmat=np.array([13., 13.]))
n.params["duration"] = 65000
# n.params["N_tp"] = 0.0
# n.params["N_rp"] = 0.0
n.params["N_pt"] = 0.
# n.params["thlm_cmat"] = np.array([0.0])
n.params["sigma_ou"] = 0.05
n.params["ctx_thlm_delay"] = 20.
```

```python
n.run()
```

```python
plt.plot(n.t, n.rates_exc[:-1].T)
plt.ylabel("ALN cortex firing rate [Hz]")
plt.xlabel("time [ms]")
plt.twinx()
plt.plot(n.t, n.rates_exc[-1].T, color="C2")
plt.ylabel("thalamus firing rate [Hz]")
# plt.xlim([10000, 20000])
plt.title("1node ALN + thalamus:: uncoupled")
```
